```python
import numpy as np
import matplotlib.pyplot as plt

#-- Generate Data -----------------------------------------
csv = np.genfromtxt('m_0.csv', delimiter=',')
row_header = csv[0,1:-1] # Azimuth list
column_header = csv[1:,0] # Frequency list
values = csv[1:, 1:-1] # Data

#-- Process Row Data for Adapting matplotlib specification --------------------------------
row_header = np.append(row_header, [360.])
azimuths, freqs = np.meshgrid(np.radians(row_header), column_header)
values = np.insert(values, -1,values[:,0], axis=1)

#-- Plot... ------------------------------------------------
v = np.linspace(0., 2.1, 200, endpoint=True) # colorbar range and display values setting (if value is out of this range, color is white)
fig, ax = plt.subplots(1,1,subplot_kw=dict(projection='polar'), dpi=800)
p = ax.contourf(azimuths, freqs, values, v, cmap=cm.jet)#plt.get_cmap('Reds')
cbar = plt.colorbar(p)

plt.show()
```
