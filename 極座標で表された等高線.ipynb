```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
import matplotlib.colors as colors
import matplotlib.cm as cm
import pandas as pd
from matplotlib.colors import Normalize
```

```python

fig2 = plt.figure(figsize=(6,6),dpi=110)
#ax2 = fig2.add_subplot(111, projection="polar")
ax2 = fig2.add_subplot(111, projection="polar")

r = np.arange(0, 370, 10)
theta = 2 * np.pi/360*r
freq=np.array([0.0445953,0.0486315,0.053033,0.0578329,0.0630672,0.0687753,0.075,0.0817881,
                  0.0891906,0.097263,0.106066,0.1156658,0.1261345,0.1375506,0.15,0.1635762,0.1783811,
                  0.194526,0.2121321,0.2313317,0.252269,0.2751013,0.3000001,0.3271524,0.3567623]) #np.arange(0, 1.0, 0.1)

wavedate=pd.read_csv("sample.csv", encoding="shift_jis",engine="python")#
wavedata=[[float(wavedate.iloc[i][j+1])for j in range(37)] for i in range(25)]#wavedate.values[0:25,1:37]
X, Y = np.meshgrid(theta, freq)
print(np.shape(X))
print(np.shape(Y))
H=wavedata#X**2+Y**2
v = np.linspace(0., 40, 400)#, endpoint=True)
ctf=ax2.contour(X, Y, H,levels=v,cmap=cm.jet,linewidths=0.7,vmax=40)#, norm=Normalize(vmin=0, vmax=40)
plt.colorbar(ctf, pad=0.1,orientation="vertical")#colb = , 
#colb.set_clim(0,40)
ax2.set_theta_zero_location("N")#JWA
ax2.set_theta_direction(-1)#
ax2.set_rlabel_position(120)#
#ax2.set_rlabel_position(-2)#

plt.show()
print("Target")
```
